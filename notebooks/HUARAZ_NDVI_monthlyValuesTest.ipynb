```python
mf = ffile['NDVI_QA']#[(ffile['time.year']>=2000) &  (ffile['time.year']<=2005)]
```

```python
mf.values[mf.values<0] = np.nan
```

```python
mf['ymonth'] = ('time', [str(y)+'-'+str(m).zfill(2) for (y,m) in zip(mf['time.year'].values,mf['time.month'].values)])

grouped='ymonth'

valid_days = mf.groupby(grouped).count(dim='time') # number of valid days per month
minus =  mf.groupby(grouped).mean(dim='time')
```

```python
arr = minus.values*0
arr[(valid_days.values<1)] = np.nan
# minus.values = arr

dso = minus - arr
```

```python
dso.isel(ymonth=0).plot()
```

```python
dso.ymonth[0:6]
```

```python
times = []
for d in dso['ymonth']:
    
   # ipdb.set_trace()
    
    year = int(str(d.values)[0:4])
    month = int(str(d.values)[5:7])
    
    times.append(datetime.datetime(year,month, 15))    
```

```python
dsout = dso.assign_coords({'ymonth' : times})
```

```python
dsout = dsout.rename({'ymonth' : 'time'})
```

```python
dsout.to_netcdf('/home/ck/DIR/cornkle/data/HUARAZ/Lorenz_NDVI/RioSanta_NDVI_monthly.nc')
```
