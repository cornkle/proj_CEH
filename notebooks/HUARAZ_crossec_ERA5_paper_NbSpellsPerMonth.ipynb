```python
eq = pkl.load(open("/home/ck/DIR/cornkle/data/HUARAZ/saves/equatorial_dw_spellDates_0.15perc_v2.p", "rb"))
tr = pkl.load(open("/home/ck/DIR/cornkle/data/HUARAZ/saves/transition_dw_spellDates_0.15perc_v2.p", "rb"))
ce = pkl.load(open("/home/ck/DIR/cornkle/data/HUARAZ/saves/central_dw_spellDates_0.15perc_v2.p", "rb"))
so = pkl.load(open("/home/ck/DIR/cornkle/data/HUARAZ/saves/southern_dw_spellDates_0.15perc_v2.p", "rb"))
```

```python
lab = ['Equatorial', 'Transition', 'Central', 'Southern']
```

```python
# Make a multiple-histogram of data-sets with different length.
f = plt.figure(figsize=(12,3), dpi=200)
ax = f.add_subplot(121)

n_bins=12
stag = 'dry3'
label=lab
x_multi = [eq[stag].month.values, ce[stag].month.values, tr[stag].month.values, so[stag].month.values]
ax.hist(x_multi, n_bins, histtype='bar', label=lab, density=True, align='left')
ax.set_xticks(np.arange(1,13))
ax.set_title('3-day dry spells')
plt.legend()

ax = f.add_subplot(122)

n_bins=12
stag = 'wet3'
x_multi = [eq[stag].month.values, ce[stag].month.values, tr[stag].month.values, so[stag].month.values]
ax.hist(x_multi, n_bins, histtype='bar', density=True, align='left')
ax.set_xticks(np.arange(1,13))
ax.set_title('3-day wet spells')
```
