```python
import xarray as xr
import matplotlib.pyplot as plt
import numpy as np
from utils import u_arrays as ua
import cartopy 
import cartopy.crs as ccrs
%pylab inline
```

```python
files = ua.locate(".nc", '/users/global/cornkle/TRMMfiles')
tl = []
pl = []
for f in files:
    dic = xr.open_dataset(f)
    if not np.isfinite(np.array(dic['t'].values)).any():
        #print('not in')
        continue
    if np.nanmin(dic['p'].values)<0:
        print(f)
    #t = np.array(dic['t'].values)[np.isfinite(np.array(dic['p'].values))]
    #p = np.array(dic['p'].values)[np.isfinite(np.array(dic['p'].values))]
    t = np.percentile(np.array(dic['t'].values)[np.isfinite(np.array(dic['p'].values))], 10)
    p = np.percentile(np.array(dic['p'].values)[np.isfinite(np.array(dic['p'].values))], 90)
    tl.extend(t)  
    pl.extend(p)
```

```python
tl = np.array(tl)
pl = np.array(pl)
```

```python
plt.scatter(tl[pl>=0],pl[pl>=0])
```
